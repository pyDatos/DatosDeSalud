```python
%matplotlib nbagg
from openpyxl import load_workbook 
import matplotlib.pyplot as plt

#Se abre el archivo:
libro = load_workbook(filename = 'Vigilancia_Rutinaria/Vigilancia rutinaria por eventos Departamental 2012.xlsx',read_only=True)

#Hoja de cálculo dentro del libro:
hoja_cal = libro['Vigilancia envento Dep']

#Acá debería imprimir "BOGOTA"
print(hoja_cal['D33'].value)
```

```python
Datos=[]
for row in hoja_cal['E33:BD33']:
    for cell in row:
        Datos.append(cell.value) 

plt.figure()        
plt.plot(range(len(Datos)),Datos);
```
